```python
gas = pd.read_csv('gas_prices.csv')


plt.title("Gas Prices over Time", fontdict = {'fontweight': 'bold', 'fontsize': 20})

#plt.plot(gas.Year, gas.UK,'o-', label = 'United Kindom')
#plt.plot(gas.Year, gas.USA,'o-', label = 'United State')
#plt.plot(gas['Year'], gas['South Korea'],'o-', label = 'South Korea')


## For all the countries
#for country in gas:
    #if country != 'Year':
        #plt.plot(gas['Year'], gas[country], marker = '.', label = country)

        
## For specific country
countries_to_look = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
    if country in countries_to_look:
        plt.plot(gas['Year'], gas[country], marker = '.', label = country)
        
# X axis ajestment        
plt.xticks(gas.Year[::3]) # [::3] mean start with first one and end with last year and skip by 3.

# Labels
plt.xlabel('Year')
plt.ylabel('US Year')

# Legend
plt.legend(title = 'Countries',bbox_to_anchor=(1, 1), loc='upper left')
plt.savefig('Gas_price',dpi = 200)
plt.show()
```

### Load Fifa data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

#### Histrograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color = '#378f2e')
plt.xticks(bins)

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Skill Level')

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

label = ['Left', 'Right']

plt.pie([left, right], labels = label, autopct = '%.2f %%')
plt.title('Foot Preferrence of FIFA')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight <125].count()[0]
light_weight = fifa.loc[(fifa.Weight >- 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >=150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight <200)].count()[0]
heavy = fifa[fifa.Weight >=200].count()[0]

label = ['Under 125','125 - 150', '150 - 175', '175 - 200', '200 Above' ] 
explored = (0, 0, 0, 0, .4)

plt.pie([light, light_weight,medium, medium_heavy, heavy], labels = label, autopct = '%.2f %%', pctdistance = 0.7, explode = explored)
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

plt.figure(figsize = (4,4))
plt.style.use('default')

label = ['FC Barcelona', 'Real Mardrid']
boxes = plt.boxplot([barcelona, madrid], labels = label, patch_artist=True, medianprops={'linewidth': 2})


for box in boxes['boxes']:
    # Set edge color
    box.set(color ='#356394', linewidth = 2 )
    
    # Change Fill Color
    box.set(facecolor = '#b1bdb6') # make sure to pass the argument that patch_artist.


plt.title('FIFA Team Comparison')
plt.show()
```
