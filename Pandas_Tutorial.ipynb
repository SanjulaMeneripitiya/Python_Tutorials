# Loading data into Pandas

```python
import pandas as pd

df = pd.read_csv('employee-data (1).csv')
print(df.head(3)) # First 3 Data
print(df.tail(3)) # Last 3 Data

# df = pd.read_excel() for excel files
# df = pd.read_csv('', delimiter = '') for txt files
```

# Reading Data in Pandas

```python
# Reading Headers
df.columns

# Reading each columns
print(df['first_name'][0:5])
print(df[['first_name', 'last_name', 'gender']][0:3])

# Reading each rows
print(df.iloc[0])
# for index,row in df.iterrows():
#     print(index,row)
print(df.loc[df['gender'] == 'M'][0:5])

# Reading a specific location (Row,Column)
print("first Name of employee 10002: ",df.iloc[2,1])
```

# Sorting/Describing Data

```python
df.describe() # summary statistics
```

```python
df.sort_values('first_name')[0:5] # Ascending
df.sort_values('first_name', ascending=False)[0:5] # Descending

df.sort_values(['salary'],ascending= True)[0:5] # Low salary
df.sort_values(['salary'],ascending= False)[0:5] # High salary

df.sort_values(['first_name','salary'],ascending= [1,0])[0:5] # first one(first_name) to be ascending second one(salary)
#to be decending
```

# Making changes to the Data

```python
 # drop column
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]]]
df.head()
```

# Saving our Data (Exporting into Desired Format)

```python
df.to_csv('Modified.csv', index= False)
df.to_excel('Modified.xlsx', index= False)
df.to_csv('Modified.txt', index= False ,sep = '\t')
```

# Filtering Data

```python
new_df = df.loc[(df['gender'] == 'M') & (df['salary'] >60000)] # and
#df.loc[(df['gender'] == 'M') | (df['salary'] >60000)] # or

new_df.to_csv('filtered.csv')
new_df = new_df.reset_index(drop = True)
new_df
```

```python
df.loc[df['title'].str.contains('Senior')]
df.loc[~df['title'].str.contains('Senior')] # not contains senior
```

```python
import re

df.loc[df['title'].str.contains('Staff|Engineer',regex = True)]
```

```python
# start with A
df.loc[df['first_name'].str.contains('^a[a-z]*',flags=re.I,regex = True)][0:5]
```

# Conditional Changes

```python
df.loc[df['gender'] == 'M','gender'] = 'Male'
df.loc[df['gender'] == 'F', 'gender'] = 'Female'
df
df.to_csv('Modification_2.csv', index = False)
```
